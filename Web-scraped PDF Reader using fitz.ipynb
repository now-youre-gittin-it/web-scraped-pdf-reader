```python
#pip install urllib
from urllib.request import urlretrieve
urlretrieve("http://www.zen134237.zen.co.uk/American_Superstore_1x01_-_Pilot.pdf", "Superstore_pilot.pdf")

```

```python
import fitz  # this is pymupdf

with fitz.open("Superstore_pilot.pdf") as doc:
    text = ""
    for page in doc:
        text += page.get_text()

print(text)
```
